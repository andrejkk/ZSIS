## ■ Opis teme: maksimalen prenos moči

Teorem o maksimalnem prenosu moči podaja pogoje, pri katerih v sistemu v izmeničnem stacionarnem stanju dosežemo Maksimalen prenos delovne moči z generatorja na breme. 

Situacijo v splošnem predsavimo z gemeratorskim dvopolom G in bremenskim dvopolom B. Oba četveropola ločeno poensotavimo po Theveninovem teoremu in dobimo situacijo na sliki 1, kjer je notranja impedanca genratorja označena z $Z_g$ in bremena u $Z_b$.

Teorem o maksimalnem prenosu moči pove, da breme $Z_b$ prejema maksimalno delovno moč le v primeru, ko velja 
$$ Z_b = Z_g^*. $$

Spomnimo, da problem maksimalnega prenosa moči nastopi ob vsakem sklopu dveh poti kot so ...

Odboj signala od meje sklenitve (ki se prišteje izvornemu signalu in povzroči poškodbo signala) nastopi natanko tedaj ko na sklenitvi ni izpolnjen pogoj maksimalnega prenosa moči. Torej je Maksimalen prenos moči nujen ne le zaradi izkoristek, ampak tudi zaradi prenosa signala brez poškodb.

## ■ Teorem: Maksimalen prenos moči

#### Teroem:
Za situacijo na sliki 1 (generatorski in bremenski četveropol sta predstavljena s Theveninovim ekvivalentnim pogojem) je največji prenos delovne moči na breme dosežen pri pogoju
$$ Z_b = Z_g^* $$
V realni obliki sta to dve enačbi $R_b=R_g$ in $X_b=-X_g$.

Če so pri prolagajanju omejitve, potem

i. Če upornosti bremena $R_b$ ni mogoče prilagoditi na generator, reaktanco prilagodimo po enaki realni enačbi:
$$ X_b = -X_g $$

ii. Reaktance bremena ni mogoče prilagoditi na generator, za breme izberemo:
$$ R_b = |Z_g| $$


#### Opombe:

Ko je breme prilagojeno, torej $Z_b = Z_g^*$, breme prejema razpoložljivo moč $P_b = P_g=\frac{|U_g|^2}{4 R_g}$. Ko breme ni prilagojeno, se del razpoložljive moči odbije od meje in breme prejme le 
$$ P_b = P_g - P_o. $$
