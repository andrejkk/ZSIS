## ■ Opis teme: Fourierjeva trigonometrična vrsta

Fourierjeva trigonometrična vrsta je predstavitev periodičnega signala kot vsota harmoničnih funkcij (harmonskih komponent), kar pomeni preslikavo iz časovnega v frekvenčni prostor. Predtavitev signala v frekvenčnem prostoru imenujemo spekter in je sestavljen in amplitudnega in faznega spektra.

## ■ Teorija

Fourierjeva trigonometična vrsta obstaja za peridične signale $x(t)$ in ima dve obliki,
$$ \hat{x}(t) = a_0 + \sum_{n=1}^\infty (a_n \cos(n\omega_1 t) + b_n \sin(n\omega_1 t)) = c_0 + \sum_{n=1}^\infty c_n \cos(n\omega_1 t + \varphi_n). $$
Signal je periodičen, če obstaja taka konstanta $T$, da je $x(t+T)=x(T)$ in perioda signala je najmanjše tako število. 

Kot kaže enačba na desni, je periodični signal $x(t)$ mogoče zapisati kot vsoto harmoničnih signalov $c_n \cos(n\omega t + \varphi_n)$, ki jih imenjujemo harmonske komponente. Pri indeksu $n$ je to $n$-ta harmonska komponenta, tu je $n=0, 1, 2, \ldots$. Osnovna krožna frekvenca je $\omega_1=\frac{2\pi}{T}$ in je najmanjša krožna frekvenca v signalu (če ne štejemo enosmerne komponente s frekvenco 0). Krožna frekvenca $n$-te harmonske komponente je $n\omega_1$ in je celoštevilski večkratnik osnovne krožne frekvence. Torej peridični signal vsebuje samo celoštevilske večkratnike osnovne frekvence in nobene druge frekvence. 

Amplituda $n$-te krožne frekvence je $c_n$ in nam pove koliko moči prenaša ta komponenta. Zbirko amplitud $\{c_n:n\}$ imenujemo amplitudni spekter.

Faza $n$-te harmonske komponente je označena z $\varphi_n$, zbirko faz $\{\varphi_n\}$ imenujemo fazni spekter signala. 

Spekter signala določimo preko enačb za koeficinete $a_0$ ter $a_n$ in $b_n$. Enosmerna komponenta je
$$ a_0 = \frac{1}{T} \int_{\tau}^{\tau+T} x(t) dt $$
in za $n=1, 2, \ldots$ še
$$ a_n = \frac{2}{T} \int_{\tau}^{\tau+T} x(t) \cos(n\omega_1 t) dt, \qquad b_n = \frac{1}{T} \int_{\tau}^{\tau+T} x(t)\sin(n\omega_1 t) dt. $$
Od tod določimo amplitudo in fazo
$$ c_n = +\sqrt{a_n^2 + b_n^2}, \qquad \varphi_n = Arg(a_n - j b_n). $$
