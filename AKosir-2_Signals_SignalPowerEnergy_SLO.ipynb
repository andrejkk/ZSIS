## ■ Opis teme: Signal, moč in energija

Signal v abstraktni obliki je vsaka entiteta, ki prenaša informacijo. V elektotehiški praksi pa signali predstavljajo električni količini napetosti (potenciala) in toka. Fizikalno realnost signaov opišemo s časovnimi funkcijami, to je s funkcijami, odvisnimi od časa. Analiza električ nih sistemov je pokazala, da je ta neobhodno povezana s signali in njihovo predstavitvijo. Izkazalo se je, da predstavitev signalov s časovnimi funkcijami ni najbolj učinkovita predstavitev in vpeljali so abstraktne predstavitve v frekvenčni in Laplaceovi domeni. Tu izraz domena pomeni množico neodvisnih spremenljivk signala. Časovno domena je množica realnih števil $R$, harmonične signale bomo predsaili s kazalci v kompleksni domeni (konmpleksna števila), periodične v domeni diskretnih frekvenc, itd. 

Signal v časovni domeni predstavimo s funkcijo na realni osi, npr. $2 \cos(6t + \pi/4) V$. Namen tega odseka je vpeljava moči in energije signalov, ki je bistvena za razumevanje električnih sistemov.

## ■ Teorija

#### Signali v časovni domeni

Signal v časovni domeni predstavimo s časovno funkcijo $x: R \to R$, kjer $x(t)$ predstavlja napetost ali tok. Pomembne skupine signalov so konstanvni signal $x(t)=A$, enotina stopnica $1(t)$, eksponentno upadanje $x(t) = A e^{-\delta t}$,  harmonični signal $x(t)=A\cos(\omega t + \varphi)$. 



#### Moč

Trenutna moč je definirana na dvopolu (slika ) in je dana z 
$$  p(t) = u(t) i(t).   $$
Izvor te definicije bomo pojasnili v naslednji točki ob vpeljavi energije. Trenutna moče je dana v Watih $[W]$.

Skozi prakso se je uveljavil pojem delovne moči, ki ga lahko smiselno vpeljemo le za harmonične signale. Taki signali nastopajo pri prenosu električne moči. Če opazujemo trenutno moč harmoničnega signala, dobimo periodični signal, ki se torej skozi čas spreminja,
$$ p(t) = I_0\cos(\omega t + \varphi_i) U_0\cos(\omega t + \varphi_u). $$

Pri tem smo upoštevali, da sta krožni frekvenci toka in napetosti enaki (utemeljitev sledi kasneje v okviru harmoničnega stacionarnega stanja), zato imata skupno periodo $T$. Glede na relevantnost v praksi, delovno moč definiramo kot moč, ki jo tak signal opravlja v praksi - to je povprečna moč po eni periodi, ki je enaka
$$ P = \frac{1}{T} \int_0^T I_0\cos(\omega t + \varphi_i) U_0\cos(\omega t + \varphi_u) dt 
 = \frac{U_0 I_0}{2} \cos(\varphi_u - \varphi_i). $$

Opazimo, da je delovna moč neodvisna od frekvence harmoničnega signala $\omega$. 



#### Energija

Energija, ki v časovnem intervalu $[t_1, t_2]$, steče v dvopol, je dana z 
$$ W(t_1, t_2) = \int_{t_1}^{t_2} u(t) i(t) d t. $$
Ozadje za tako definicije izhaja iz osnovne definicije dela in moči v fiziki, kjer je energija v najpreprostejši obliki dana z enačbo $F s$, kjer je $F$ konstantna sila, ki jo telo premaguje na razdalji $s$. Pri dvopolu napetost, ki je razlika potencialov, poganja naboj preko električnega polja, kjer čuti silo. Natančen razmislek vodi v enačbo električne energije. 

#### Energija abstraktnega signala

Pri analizi sistemov in signalov se izkaže potreba, da bi energijo definirali za en sam signal $x(t)$, enačba za energijo pa zahteva signal napetosti in toka. Težavo rešimo tako, da energijo abstraktnega signala definiramo kot energijo, ki steče v dvopol, ki ga sestavlja le upor z upornostjo $R=1\Omega$. 
Tedaj dobimo
$$ W(t_1, t_2) = \int_{t_1}^{t_2} R i(t)^2 d t = \int_{t_1}^{t_2} \frac{u(t)^2}{R} d t $$
in od tod za $R=1\Omega$ energijo abstraktnega signala $x(t)$

$$ W(t_1, t_2) = \int_{t_1}^{t_2} x(t)^2  d t. $$
