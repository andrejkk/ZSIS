## ■ Opis teme: Lastni odziv vezja

Lastni odziv vezja je del odziva vezja, ki je posledica energije, shranjene v reaktivnih elelemntih. Skupaj z vsiljenim odzivom , ki je posledica neodvisnih virov po času $0$, sestavlja celoten odziv sistema.

Lastni odziv je v celoti dan z ničlami karakterističnega polinoma $P(p)=0$, ki jih označimo z $p_1, \ldots, p_n$, kjer je $n$ red sistema.

Zanimive je, da so možni le štirje različni tipi lasntega odziva linearnega sistema, ki so 
1. Nedušeno nihanje: v tem primeru v vezju ni nobenih izgub, to je nobenih uporov, le L in C elementi. Tako to ni realističen primer. Ničle sistema so v tem primeru cisto imaginaren. 
2. Dušeno nihanje
3. Kritično dušeno nihanje
4. Nadkritično dušeno nihanje

## ■ Teorija

#### Enačba sistema

Diferencialno enačbo zapišemo s simboličnim sistemskim operatorjem kot 
$$ P(D) y(t) = Q(p) x(t). $$

#### Lastni odziv je homogeni del rešitve 

Homogeni del je glede tipa odziva določen z ničlami karakterističnega polinoma 
$$
P(p) = 0: \; p_1, \ldots, p_n,
$$
kjer vsaka ničla $p_i$ prispeva člen 
$$ 
K e^{p_i t}.
$$ 
Če je 
- realni del ničle $Re(p_i)=0$, je njen prispevek k lastnemu odzivu nedušeno nihanje;
- ničla kompleksna prvega reda ($Re(p_i)<0$,$Im(p_i)\ne 0$), je prispevek dušeno nihanje
- ničla realna višjega reda, je prispevek kritično dušenje. 
- ničla realna prvega reda, je prispevek nadkritično dušenje.
