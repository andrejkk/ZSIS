## ■ Opis teme: resonanca in kvaliteta

Resonanca je pojav, ko sistem v izmeničnem stacionarnem stanju pri dani frekvenci doseže maksimalno amplitudo odziva, pri čemer je amplituda vzbujanja konstanta. Tako frekvenco imenujemo resonančna frekvenca. Pri tem je zanimivo, da ima linearno vezje lahko več različnih resonančnih frekvenc, ki so različno izrazite. To izrazitost merimo s kvaliteto resonance.

## ■ Teorija

#### Resonančna frekvenčna

Napetostna resonančna frekvenca $f_0$ je frekvenca, pri kateri ima amplituda napetiosti dvopola $|U|$ v izmeničnem stacionarnem stanju v odvisnnosti od frekvence lokalni maksimum, pri čemer je amplituda toka $|I|$ na dvopolu konstantna, glejte sliko. 

Napetost in tok dvopola sta povezana po enačbi $U = Z_N(j\omega) I$, kjer je $Z_N(\omega)$ notranja impedanca tega dovopola. 
Neposredno določanje resonančne frekvence bi zahtevalo iskanje ekstremov funkcije $|Z_N(\omega) I|$, torej računanje prvih in drugih odvodov itd. Zato v praksi uporabimo približno imitančno formulo, ki je
$$ B(j\omega_0) = 0, $$
kjer je $B(\omega)$ notranja susceptanca dvopola. Enačba je dovolj enostavna in daje uporabne rezultate, njene rešitve so napetostne resonanče frekvence dvopola (navadno je rešitev tudi $\omega_0=0$, kar ni smiselna rešitev).      

Na podoben način definiramo tokovo resonančno frekvenco, ki je frekvenca, pri kateri ima amplituda odziva $|I|$ lokalni maksimum, pri čemer je amplituda vzbujanja $|U|$ konstantna. Tudi za te resonančne frekvence imamo približno imitančno formulo, ki je
$$ X(j\omega_0) = 0, $$
kjer je $X(\omega)$ notranja reaktanca dvopola.

Izkaže se, da so za poljuben realen dvopol (dvopol z upori) položaji vseh tokovih resonančnih frekvenc enake napetostnim. V $LC$ četveropolu pa si te frekvenco sledijo izmenično.  
 
#### Kvaliteta resonančne frekvence

Kvaliteta meri izrazitost resonačne frekvence, pri čemer je bolj izrazita resonanca bolj kvalitetna. Definiramo jo kot razmerje
$$ Q(\omega_0) = \frac{f_0}{B}, $$
kjer je $\omega_0 = 2\pi f_0$ krožna resonančn frekvenca in $B$ je frekvenčni pas $B = |f_2-f_1|$. Mejni frekvenci $f_1$ in $f_2$ sta dani z upadom maksimalne amplitude $A$ na $A/\sqrt{2}$ (pred f_0 odčitamo $f_1$, po $f_0$ pa $f_1$). Razlog za izbiro faktorja $1/\sqrt{2}$ oz. $-3 dB$ je povezan s stabilnostjo sprejemnika z več kanali in ga tu ne argumentiramo. 

Neposredni izračun kvalitete je precej zapleten in tudi tu uporabimo imitančne formule 
$$ Q(\omega_0) = \frac{X_L}{R} = -\frac{B_C}{G}. $$
Tu je $X_L$ induktivni del reaktance in $B_C$ je kapacitivni del susceptance. Te imitančne formule so enostavne samo navidez, saj je lahkp določanje induktivnege in kapacitivnega dela precej zapleteno. 

Zato je bolj uporabna energijska formula za kvaliteto
$$ Q(j\omega) = \omega_0 \frac{\sum_L W_L(\omega_0)}{\sum_R P(\omega_0)} 
  = \omega_0 \frac{\sum_C W_C(\omega_0)}{\sum_R P(\omega_0)}, $$ 
kjer je $\sum_R P(\omega_0)$ vsota delovnih moči vseh uporov, $\sum_L W_L(\omega_0)$ vsota maksimalnih energij tuljav in $\sum_C W_C(\omega_0)$ vsota maksimalnih energij kondenzatorjev. Pri tem velja, da je v vsaki resonančni frekvenci vsota maksimalnih energij tuljav enak vsoti vsota maksimalnih energij kondenzatorjev.
