## ■ Opis teme: linearnost in časovna invariantnost

Sistem je linearen, če je odziv na linearno kombinacijo dveh vzbujanj enak linearni kombinaciji teh odzivov (glejte definicijo). Najpreprostejša linearna kombinacija je vsota. Če je vzbujanje tok, potem je odziv na vzbujanje z vsoto dveg tokov ${\cal H}(i_1(t)+i_2(t))$ enak vsoti odzivov na posamezna tokova ${\cal H}(i_1(t))+{\cal H}(i_2(t))$. Linearnost oz. nelinearnost sistema bistveno vpliva na vse faze analize, od postavljanja enačb sistema do razumevanja povezave med odzivom in vzbujanjem. Izkaže se, da so linearni natanko sistemi, ki jih sestavljajo linearni elementi. Linearni elementi so upor, tuljava, kondenzator in krmiljen vir (lineariziran model sklopov ipd).

Sistem je časovno invarianten če je njegov odziv na dano vzbujanje neodvisno od absolutnega časa ko se to vzbujanje pojavi. Torej zakasnjeno vzbujanje $x(t-\tau)$ za čas $\tau$ da enak odziv kot brez zakasnitve $y(t)$, le z enako zakasnitvijo, torej $y(t-\tau)$. Gre za to, da se sistem skozi čas ne spreminja. Preprost časovno neinvarianten sistem je sistem s temperaturno odvisnim uporom $R = R(T)$, ki se v vezju segreva in tako vezja spreminja svoje lastnosti. 

Sistem abstraktnmo predstavimo z 
$$ y(t) = {\cal H}(x(t)) $$

## ■ Definicija: Linearni sistem

Sistem ${\cal H}$ je linearen, če je 
- Homogen: za poljubno konstanto $a$, signal $x(t)$ in čas $t$ velja 
$$ {\cal H}(a x(t)) = a {\cal H(x(t))} $$
- Aditiven: za poljubna signala $x_1(t)$, $x_2(t)$ in čas $t$ velja
$$ {\cal H}(x_1(t) + x_2(t)) = {\cal H(x_1(t))} + {\cal H(x_2(t))} $$

Ekvivalentno, če za poljubni konstanti $a_1$, $a_2$, signala $x_1(t)$, $x_2(t)$ in čas $t$ velja
$$ {\cal H}(a_1 x_1(t) + a_2 x_2(t)) = a_1 {\cal H(x_1(t))} + a_2 {\cal H(x_2(t))} $$

## ■ Definicija: Časovno invarianten sistem

Sistem ${\cal H}$ je časovno invarianten, če za poljuben signal $x(t)$ in poljubno zakasnitev $\tau$  velja
$$ y(t-\tau ) = {\cal H}(x(t-\tau)), $$
kjer je $y(t) = {\cal H}(x(t))$ odziv na nezakasnjeno vzbujanje.

### Primer nelinearnega sistema: dioda

Karakteristika Shocklijeve diode je dana z 
$$ i(t) = I_s \left(e^{\frac{u(t)}{n V_T}} - 1\right), $$
kjer je $I_s$ tok saturacije (med $10^{-12}$ in $10^{-9}$), faktor idealnosti $n$ (med 1 in 2) ter termalna napetost (pribl. $25,8 mV$).

Za model diode 1N4148 je $I_s = 2\cdot 10^{-9}$, $n=1.7$ in $V_T=25.85$. 



```python
import numpy as np
import matplotlib.pyplot as plt

def system_1N4148(u):
    I_s = 2e-9
    n = 1.7
    V_T = 25.85e-3

    return I_s*(np.exp(u / (n*V_T)) - 1)

u = np.arange(0, 1, 0.01)
plt.plot(u, system_1N4148(u))
plt.xlabel('u[V]')
plt.ylabel('i[A]')
```
