## ■ Opis teme: karakteristična impedanca

Karakteristična impedanca je impedanca, ki se kot breme simetričnega četveropola preslika v njegovo vhodno impedanco. Torej če četveropol A obremenimo (priključimo na drugi vhod) z njegovo karakteristično impedanco $Z_k$, potem je njegova vhodna impedanca (na prvem vhodu) enaka tej karakteristični impedanci $Z_1 = Z_v = Z_k$. Od tod izhaja njena definicijska enačba. 

Definicija karakteristične impedance izhaja iz pogojev o maksimalnem prenosu moči, ki naj jih četveropol ob vstavitvi med generator in breme ohrani. Če je breme $Z_b$ prilagojeno na generator pred vstavitvijo četveropola (glejte sliko 1), naj bo prilagojeno po vtavitvi četveropola. Da bo to držalo, mora po vstavitvi četveropola generator čutiti enako uporanost kot pred vstavitvijo - prej je čutil $Z_b$, po vstavitvi pa čuti $Z_v(Z_b)$ in ta mora torej biti enaka $Z_b$. Tu je $Z_b \to Z_v(Z_b)$ preslikava impedance bremena $Z_b$ na vhod četveropola. Tako mora veljati 
$$ Z_b = Z_v(Z_b). $$
To bo izpolnjeno če je $Z_b$ enak karateristični impedanci četveropola A in od tod izhaja definicija karakteristične impedance.

## ■ Načini določanja karakteristične impedance

Karakteristično impedanco lahko izračunamo na več načinov. Prvi način je z reševanjem definicijske enačbe 
$$ 
Z_k = Z_v(Z_k), 
$$
drugi način je na podlagi četveropolnih parametrov. Za izbran nabor četveropolnih parametrov lahko izpeljemo enačbo kaakteristične impedance, npr. za impedančne parameter velja 
$$ 
Z_k = \pm\sqrt{\triangle (z)},
$$

Poleg naštetih načinov velja tudi enačba
$$ 
Z_k = \pm\sqrt{Z_{vho} Z_{vhk}}, 
$$
kjer je $Z_{vho}$ vhodna impedanca pri odrtem drugem vhodu, $Z_{vhk}$ pa vhodna impedanca pri kratko sklenjenem drugem vhodu. 

Predznak pri kvadratnih korenih določimo tako, da je $Re(Z_k)\geq 0$.

## ■ Simulacija 1: Določanje karakteristične impedance

Z uporabo relacije $Z_v = Z_v(Z_b)$ lahko določimo karakteristično impedanco četveropola, in to na dva načina. 

Prvi način:  po definicijski enačbi $Z_k = Z_v(Z_k)$, kjer torej izberemo tako breme, da se preslika vase na vhod. 

Drigi način: z izbiro bremena $Z_b=0\Omega$ odčitamo $Z_{vhk}$ in z izbiro bremena $Z_b=\infty\Omega$ še $Z_{vho}$. Nato po enačbi $Z_k = \pm\sqrt{Z_{vho} Z_{vhk}}$ s korenjenjem kompleksnih števil določimo $Z_k$.
